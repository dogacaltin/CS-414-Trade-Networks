```python
countries1 = df_trades["importer1"].unique()
dicti = {}
for state in countries1:
    countries2 = df_trades[df_trades["importer1"]==state]["importer2"].unique()
    dicti[state] = len(countries2)
dicti = dict(sorted(dicti.items(), key=lambda item: item[1], reverse=True))
for i in dicti:
    print(i, dicti[i])
```
